**USER DEFINED FUNCTION**



```
def function(m1,m2):
  sum=m1+m2
  return sum
m1=int(input("Enter the number: "))
m2=int(input("Enter the number: "))
result=function(m1,m2)
print("The sum is ",result)
```


**FUNCTION WITH MULTIPLE PARAMETERS**

```
def divisor(x,y,m):
    if x%m==0 and y%m==0:
        print("%d & %d are divisible by %d"%(x,y,m))
    elif x%m==0:
        print("only %d is divisible by %d"%(x,m))
    elif y%m==0:
        print("only %d is divisible by %d"%(y,m))
    else:
       print("none of %d & %d is divisible by %d"%(x,y,m))

divisor(10,20,5)
divisor(12,20,5)
divisor(10,22,5)
divisor(20,45,3)
```

**FUNCTION WITHOUT PARAMETER**

```
def add():
  m=2+3
  return m
m1=add()
print(m1)
```

**Function with default parameter value**

```
def check(a=0):
    if a==0:
        print("Zero")
    elif a>0:
        print("Positive")
    else:
        print("Negative")

check()     
check(-8)
```

**Global scope**

```
def global_scope():
    global x
    x=50

global_scope()
print(x)
```
